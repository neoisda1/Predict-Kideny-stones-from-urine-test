```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
#importing data
data_set=pd.read_csv('urine.csv')
X=data_set.iloc[:,[0,1,2,3,4,5]].values
y=data_set.iloc[:,[6]].values
#flattening the array
y = y.ravel()
#splitting the data set into training and test
from sklearn.model_selection import train_test_split
X_train,X_test, y_train, y_test=train_test_split(X,y,test_size=0.2,random_state=0)
#feature scaling
from sklearn.preprocessing import StandardScaler
SC=StandardScaler()
X_train=SC.fit_transform(X_train)
X_test=SC.transform(X_test)
#create the model for logistic or sigmoid regression
from sklearn.linear_model import LogisticRegression
classifier=LogisticRegression(random_state=0, solver ='lbfgs')
classifier.fit(X_train, y_train.astype('int'))
#predict on test dataset
y_pred=classifier.predict(X_test)
#evaluate accuracy
from sklearn.metrics import confusion_matrix
CM = confusion_matrix(y_test,y_pred)
CM
#test the prediction
print(classifier.predict(X_test[0].reshape(1,-1)))
```
